```python
# Step 1: Import Required Library
# This step imports the pandas library to work with DataFrames.
import pandas as pd
```

```python
# Step 2: Create the Student DataFrame
# This step creates a pandas DataFrame using the provided student data.
data = {
    "StudentID": [101,102,103,104,105,106,107,108,109,110],
    "Name": ["Alice","Bob","Charlie","Diana","Ethan","Fiona","George","Hannah","Ivan","Julia"],
    "Age": [20,21,19,22,20,21,23,19,22,20],
    "Marks": [88,75,92,65,78,85,70,95,60,80],
    "Grade": ["A","B","A","C","B","A","B","A","C","B"]
}

df = pd.DataFrame(data)
df
```

```python
# Step 3: Save the DataFrame as CSV
# This step saves the DataFrame as a CSV file named students.csv.

df.to_csv("students.csv", index=False)
```

```python
# Step 4: Reload the CSV into a New DataFrame
# This step reloads the saved CSV file into a new DataFrame.

df_new = pd.read_csv("students.csv")
df_new
```

```python
# Step 5: Display First 3 Rows
# This step displays the first 3 rows of the DataFrame.

df_new.head(3)
```

```python
# Step 6: Display Statistical Summary
# This step shows statistical summary of numerical columns.

df_new.describe()
```

```python
# Step 7: Explore Data Further
# This step shows number of rows and columns.

df_new.shape
```

```python
# Display Data Types
# This step shows data types of each column.

df_new.dtypes
```

```python
# Display Complete Info
# This step shows detailed information including non-null counts.

df_new.info()
```
